# Mental Health Dataset Analysis (DIU Extended Edition)

This notebook loads and analyzes the processed mental health dataset: **"Mental Health Problems of University Students (DIU Extended Edition)"**.

### Dataset Overview:
- **Base dataset:** Standardized dataset on Mental Health Problems (MHPs) of University Students (Syeed et al., 2024, Figshare DOI: [10.6084/m9.figshare.25771164.v1](https://doi.org/10.6084/m9.figshare.25771164.v1)).
- **Extended dataset:** Combined with local DIU-collected responses.
- **Variables:** 
  - Demographics (Age, Gender, University, Department, Academic Year, CGPA, waiver status)
  - PSS-10 scale (Perceived Stress Scale, items PSS1–PSS10, total value, and label)
  - GAD-7 scale (Generalized Anxiety Disorder, items GAD1–GAD7, total value, and label)
  - PHQ-9 scale (Patient Health Questionnaire, items PHQ1–PHQ9, total value, and label)

```python
import os
import pandas as pd
import numpy as np

data_path = "diu-extended-mental-health/data/Updated Processed Diu.csv"

if not os.path.exists(data_path):
    print(f"Error: '{data_path}' not found.")
else:
    df = pd.read_csv(data_path)
    print("\n--- Dataset successfully loaded! ---")
    print(f"Dimensions: {df.shape[0]} rows, {df.shape[1]} columns\n")
    
    print("--- First 5 Rows Preview ---")
    display(df.head())
    
    print("\n--- Dataset Summary ---")
    print(df.info())
```

### 1. Demographic Distribution
Let's explore the distribution of key demographic characteristics (Gender, Age group, Academic Year, and Current CGPA class).

```python
if 'df' in locals():
    demographics = ['Gender', 'Age', 'Academic_Year', 'Current_CGPA', 'waiver_or_scholarship']
    for col in demographics:
        if col in df.columns:
            print(f"\n=== Distribution of {col} ===")
            counts = df[col].value_counts(dropna=False)
            pcts = df[col].value_counts(dropna=False, normalize=True) * 100
            summary = pd.DataFrame({'Count': counts, 'Percentage (%)': pcts.round(2)})
            display(summary)
else:
    print("Dataset has not been loaded. Please run the load cell first.")
```

### 2. Mental Health Assessment Labels (Stress, Anxiety, Depression)
Let's check the distributions of severity levels for Stress, Anxiety, and Depression.

```python
if 'df' in locals():
    mental_health_labels = ['Stress Label', 'Anxiety Label', 'Depression Label']
    for col in mental_health_labels:
        if col in df.columns:
            print(f"\n=== Distribution of {col} ===")
            counts = df[col].value_counts(dropna=False)
            pcts = df[col].value_counts(dropna=False, normalize=True) * 100
            summary = pd.DataFrame({'Count': counts, 'Percentage (%)': pcts.round(2)})
            display(summary)
else:
    print("Dataset has not been loaded. Please run the load cell first.")
```

### 3. Correlation Analysis between Mental Health Scales
Let's see if there is any correlation between Perceived Stress, Generalized Anxiety, and Depression raw scores.

```python
if 'df' in locals():
    scale_values = ['Stress Value', 'Anxiety Value', 'Depression Value']
    valid_cols = [c for c in scale_values if c in df.columns]
    if valid_cols:
        print("=== Correlation Matrix ===")
        corr_matrix = df[valid_cols].corr()
        display(corr_matrix.round(4))
        
        print("\n=== Mean and Standard Deviation of Scales ===")
        display(df[valid_cols].describe().loc[['mean', 'std', 'min', 'max']])
else:
    print("Dataset has not been loaded. Please run the load cell first.")
```
